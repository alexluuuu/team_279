```python
from utils.colors import *
from utils.segmentation import *
from utils.DatasetPrep import GatherImageSet
from skimage import io
import matplotlib.pyplot as plt
```

```python
image_dir = 'sample_dataset/images/'
num_imgs, imgs = GatherImageSet(image_dir)
for i in range(num_imgs):
    image = io.imread(image_dir+imgs[i])
    segment = compute_segmentation(image, 3, clustering_fn=kmeans_fast, \
                                   feature_fn=color_position_features, scale=0.1)
    center = getCenterSegment(image, segment)
    plt.figure()
    plt.subplot(131)
    plt.imshow(image)
    plt.subplot(132)
    plt.imshow(segment, cmap='viridis')
    plt.subplot(133)
    plt.imshow(center)
    plt.show()
    if i == 1: break
```
